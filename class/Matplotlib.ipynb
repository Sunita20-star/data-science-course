```python
import matplotlib.pyplot as plt
labels=['samsung','apple','xiaomi','others']
sizes=[30,25,20,25]
plt.pie(sizes,labels=labels,autopct='%1.1f%%',startangle=90)  #autopct manage the percentage label
plt.title('smartphone market share')
plt.axis('equal')  #ensures the pie is in circle
plt.show()
```

```python
import matplotlib.pyplot as plt
streams=['science','commerce','arts','cs','others']
students=[120,80,100,60,40]
plt.figure(figsize=(8,8))
plt.pie(students,labels=streams,autopct='%1.1f%%',startangle=90)  #autopct manage the percentage label
plt.title('distribution of students in different streams')
# plt.axis('equal')  #ensures the pie is in circle
plt.show()
```

```python
import matplotlib.pyplot as plt

streams = ['science', 'commerce', 'arts', 'cs', 'others']
students = [120, 80, 100, 60, 40]
explode = [0.1, 0.1, 0.1, 0.2, 0.3]  # to separate slices for emphasis

plt.figure(figsize=(8, 8))
plt.pie(
    students,
    labels=streams,
    autopct='%1.1f%%',
    startangle=90,
    explode=explode  # ✅ apply explode here
)
plt.title('Distribution of Students in Different Streams')
# plt.axis('equal')  # optional for perfect circle
plt.show()
```

```python
import matplotlib.pyplot as plt

test_score = [65, 66, 67, 68, 69, 70, 71, 80, 90, 95, 100, 60, 70, 80, 90]

plt.hist(x=test_score, bins=6, edgecolor='black')
plt.title('Distribution of Students Test Score')
plt.xlabel('Score Range')        # optional but recommended
plt.ylabel('Number of Students') # optional but recommended
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)  # values mapped to colormap
sizes = 500 * np.random.rand(50)

plt.figure(figsize=(12, 4))

# Viridis
plt.subplot(1, 3, 1)
plt.scatter(x, y, c=colors, s=sizes, cmap='viridis', alpha=0.7)
plt.title('Viridis')

plt.subplot(1, 3, 2)
plt.scatter(x, y, c=colors, s=sizes, cmap='winter', alpha=0.7)
plt.title('winter')

# Coolwarm
plt.subplot(1, 3, 3)
plt.scatter(x, y, c=colors, s=sizes, cmap='hot', alpha=0.7)
plt.title('hot')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)   # values that will map to the colormap
sizes = 500 * np.random.rand(50)

plt.scatter(x, y, c=colors, s=sizes, cmap='viridis', alpha=0.7)
plt.colorbar(label="Color scale")  # adds color legend
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatter Plot with Viridis Colormap')
plt.show()
```
