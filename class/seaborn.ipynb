```python
import numpy as np
import seaborn as sns
sns.set(style='white') #background white
r=np.random.RandomState(10)  #seed for reproducibility
d=r.normal(size=100)  #100 random numbers from a normal distribution with mean=0 and standard deviation=1
sns.histplot(d,kde=True,color='m')
```
